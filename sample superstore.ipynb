# Numerical to Categorical 

```python
df['Category'].value_counts()
```

# Concentrating on high-sales, high-profit items within profitable categories can optimize both revenue and profit margins.

```python
sns.scatterplot(x='Sales', y='Profit', hue='Category', data=df)
plt.title('Sales vs. Profit')
plt.show()
```

# Focus on customer segments that yield higher profitability, such as 'Home Office.'

```python
sns.barplot(x='Segment', y='Profit', data=df)
plt.title('Profit by Customer Segment')
plt.show()
```

# Focusing on high-profit categories and tailoring promotions to the most profitable customer segments within those categories can maximize profitability.

```python
sns.barplot(x=df['Category'],y=df['Profit'],hue=df['Segment'])
```

# Revise the pricing strategy for "Same Day" shipping to better cover operational costs

```python
sns.barplot(x=df['Category'],y=df['Profit'],hue=df['Ship Mode'])
```

# Re-assess the cost structure of sub-categories like 'Tables' and 'Bookcases' that are running at a loss.

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Sub-Category', y='Profit', data=df)
plt.title('Profit by Sub-Category')
plt.xticks(rotation=45)
plt.show()
```
